```python
import pandas as pd
import glob
import os

# --- 1. PERSIAPAN ---
# Ambil semua file xlsx dan urutkan (A-Z)
list_files = sorted(glob.glob("*.xlsx"))

# Nama file output
output_csv = "main_data.csv"

# Cek apakah file ditemukan
if not list_files:
    print("Error: Tidak ada file .xlsx ditemukan. Cek panel Files.")
else:
    print(f"Ditemukan {len(list_files)} file. Memulai penggabungan hemat memori...\n")

    # --- 2. PROSES LOOPING (APPEND MODE) ---
    total_baris = 0

    for i, file_name in enumerate(list_files):
        try:
            print(f"[{i+1}/{len(list_files)}] Memproses: {file_name} ...", end=" ")

            # Membaca file Excel
            # Jika file sangat besar, engine='openpyxl' lebih stabil
            df = pd.read_excel(file_name, engine='openpyxl')

            # --- LOGIKA PENULISAN KE CSV ---
            if i == 0:
                # Jika ini file PERTAMA:
                # Mode 'w' (write) = buat file baru
                # header=True = sertakan nama kolom
                df.to_csv(output_csv, index=False, mode='w', encoding='utf-8')
            else:
                # Jika ini file KEDUA dst:
                # Mode 'a' (append) = tambahkan di bawah data yang sudah ada
                # header=False = JANGAN tulis nama kolom lagi
                df.to_csv(output_csv, index=False, mode='a', header=False, encoding='utf-8')

            # Hitung total baris untuk laporan
            total_baris += len(df)
            print("OK.")

            # Membersihkan variabel df dari memori agar RAM lega
            del df

        except Exception as e:
            print(f"\nGAGAL pada file {file_name}: {e}")

    # --- 3. SELESAI ---
    print("\n" + "="*40)
    print("PROSES SELESAI!")
    print(f"Total baris data: {total_baris}")
    print(f"File tersimpan sebagai: {output_csv}")
    print("="*40)
```
